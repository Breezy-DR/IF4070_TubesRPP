```
default_rules = [
    {"diagnosa":["demam", "batuk", "pilek"], 
    "penyakit": "Flu", 
    "kepastian": 0.9},
    {"diagnosa":["demam", "batuk"], 
    "penyakit": "Flu ringan", 
    "kepastian": 0.7}, 
    {"diagnosa":["pilek", "gatal"], 
    "penyakit": "Alergi", 
    "kepastian": 0.8},
    {"diagnosa":[], 
    "penyakit": "Tidak yakin, konsultasi dengan dokter", 
    "kepastian": 0.5}
]

def diagnose(gejala):
    diagnose = []
    for rule in default_rules:
        if set(gejala).issubset(set(rule["diagnosa"])):
            diagnose.append((rule["penyakit"], rule["kepastian"]))
    if not diagnose:
        diagnose.append(("Tidak yakin, konsultasi dengan dokter", 0.5))

    return diagnose

def add_rule():
    rule = {}
    print("Masukkan diagnosa baru:")
    print("Contoh: demam, batuk, pilek, gatal\n")
    new_diagnosa = input("Diagnosa Anda: ").lower().strip()
    gejala = [g.strip() for g in new_diagnosa.split(",")]
    new_penyakit = input("Penyakit: ").lower().strip()
    new_kepastian = float(input("Tingkat kepastian: "))
    rule["diagnosa"] = gejala
    rule["penyakit"] = new_penyakit
    rule["kepastian"] = new_kepastian
    default_rules.append(rule)
```

```
print("=== Sistem Pakar Diagnosis Kesehatan ===")
print("Apa yang bisa saya bantu?")

print("1. Tambahkan rule")
print("2. Mulai diagnosis")
print("3. Lihat rule")
choice = int(input("Masukkan pilihan: "))

if choice == 1:
    add_rule()
elif choice == 2:
    print("=== Mulai diagnosis ===")
    print("Masukkan gejala yang Anda alami (pisahkan dengan koma).")
    print("Contoh: demam, batuk, pilek, gatal\n")

    # User input
    user_input = input("Gejala Anda: ").lower().strip()
    gejala = [g.strip() for g in user_input.split(",")]

    # Diagnosis
    hasil = diagnose(gejala)

    # Output results
    print("\n=== Hasil Diagnosis ===")
    for penyakit, kepastian in hasil:
        print(f"- {penyakit}: Tingkat kepastian {kepastian * 100:.1f}%")
    print("\nSaran: Jika gejala memburuk, segera konsultasikan dengan dokter.")

elif choice == 3:
    print("=== Daftar rule ===")
    for rule in default_rules:
        print(f"Diagnosa: {rule['diagnosa']}")
        print(f"Penyakit: {rule['penyakit']}")
        print(f"Tingkat kepastian: {rule['kepastian']}")
        print("")
```
